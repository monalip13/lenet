```python
# Use seaborn for pairplot.
!pip install -q seaborn
# Importing necessary libraries.
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
# Make NumPy printouts easier to read.
np.set_printoptions(precision=3, suppress=True)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

```python
# load the dataset from below provided url.
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']
# Read the dataset using pandas data frame.
Raw_data = pd.read_csv(url, names=column_names,
                          na_values='?', comment='\t',
                          sep=' ', skipinitialspace=True)
```

```python
# creating a copy of the data set.
dataset = Raw_data.copy()
dataset.tail()
```

```python
# clean the data
```

```python
# check for the missing values
dataset.isna().sum()
```

```python
#drop rowss with the missing values.
dataset = dataset.dropna()
dataset
```

```python
# one-hot encode the values in the column with pd.get_dummies as origin column is categorical.

dataset['Origin'] = dataset['Origin'].map({1: 'USA', 2: 'Europe', 3: 'Japan'})
dataset = pd.get_dummies(dataset, columns=['Origin'], prefix='', prefix_sep='')
dataset.tail()
```

```python
# split the data into training and testing set.
train_dataset = dataset.sample(frac=0.8, random_state=0)
test_dataset = dataset.drop(train_dataset.index)
# Analyze the distribution and relationships between variables using pairplots
sns.pairplot(train_dataset[['MPG', 'Cylinders', 'Displacement', 'Weight']], diag_kind='kde')
```

```python
train_dataset.describe().transpose()
```

```python
# Create separate datasets for labels and features.
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('MPG')
test_labels = test_features.pop('MPG')
```

```python
train_dataset.describe().transpose()[['mean', 'std']]
```

```python
# Normalize the data to ensure consistent scaling across features.
normalizer = tf.keras.layers.Normalization(axis=-1)
```

```python

normalizer.adapt(np.array(train_features))
print(normalizer.mean.numpy())
```

```python
first = np.array(train_features[:1])

with np.printoptions(precision=2, suppress=True):
  print('First example:', first)
  print()
  print('Normalized:', normalizer(first).numpy())
```

```python
# Linear Regression with one variable
# Define the feature (independent variable) and label (dependent variable) for linear regression.
horsepower = np.array(train_features['Horsepower'])
# initiate the normalization using tf.keras.layers.Normalization and fit its state to the horsepower data
# Normalize the feature using a normalization layer.
horsepower_normalizer = layers.Normalization(input_shape=[1,], axis=None)
horsepower_normalizer.adapt(horsepower)
```

```python
# Build the Keras Sequential Model (This Model will predict MPG from horsepower.)
horsepower_model = tf.keras.Sequential([
    horsepower_normalizer,
    layers.Dense(units=1)
])

horsepower_model.summary()
```

```python
#Run the untrained model on the first 10 'Horsepower' values.
horsepower_model.predict(horsepower[:10])
```

```python
# compile the model using suitable optimizer and loss.
horsepower_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
%%time
#traing the linear regression model with one variable.
history = horsepower_model.fit(
    train_features['Horsepower'],
    train_labels,
    epochs=100,
    verbose=0,
    # Compute validation results on 20% of the training data.
    validation_split = 0.2)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
# plot the training and validation loss curves.
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [MPG]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```

```python
# evaluate the model's performance based on test results.
test_results = {}

y = test_results['horsepower_model'] = horsepower_model.evaluate(
    test_features['Horsepower'],
    test_labels, verbose=0)
print("test results:", y)
```

```python
# generate prediction using the trained model.
x = tf.linspace(0.0, 250, 251)
y = horsepower_model.predict(x)
#print("X:", x)
#print("Y:", y)
```

```python
#plot the linear regression line and actual data points on line.
def plot_horsepower(x, y):
  plt.scatter(train_features['Horsepower'], train_labels, label='Data')
  plt.plot(x, y, color='m', label='Predictions')
  plt.xlabel('Horsepower')
  plt.ylabel('MPG')
  plt.legend()
```

```python
plot_horsepower(x, y)
```

```python
def build_and_compile_model(norm):
  model = keras.Sequential([
      norm,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
  return model
```
