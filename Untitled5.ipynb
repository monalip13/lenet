```python
# Import necessary libraries
import numpy as np
import tensorflow as tf
import tensorflow.keras as K
import matplotlib.pyplot as plt
```

```python
# set random seed
np.random.seed(11)
tf.random.set_seed(11)

# Describe batch size for training
batch_size = 256
# Define maximum number of epochs for training
max_epochs = 20
# Define learning rate for optimizer
learning_rate = 1e-3
# Define momentum for optimizer
momentum = 8e-1
# Define hidden layer dimension
hidden_dim = 128
# Define original data dimension
original_dim = 784
# Load MNIST dataset
(x_train, _), (x_test, _) = K.datasets.mnist.load_data()
```

```python
# Normalize train and test data
x_train = x_train / 255.
x_test = x_test / 255.

# Convert train and test data to float32
x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)
x_train = np.reshape(x_train, (x_train.shape[0], 784))  # Reshape training data
x_test = np.reshape(x_test, (x_test.shape[0], 784))  # Reshape test data
```

```python
# Add noise to generate corrupted MNIST images with normal distribution
noise = np.random.normal(loc=0.5, scale=0.5, size=x_train.shape)
x_train_noisy = x_train + noise
noise = np.random.normal(loc=0.5, scale=0.5, size=x_test.shape)
x_test_noisy = x_test + noise
print("x_train_noise :",x_train_noisy )
print("x_test_noisy :",x_test_noisy )
```

```python
class Encoder(K.layers.Layer):
    def __init__(self, hidden_dim):
        super(Encoder, self).__init__()
        self.hidden_layer1 = K.layers.Dense(units=hidden_dim*2, activation=tf.nn.relu)  # Add an additional hidden layer
        self.hidden_layer2 = K.layers.Dense(units=hidden_dim, activation=tf.nn.relu)
        self.hidden_layer3 = K.layers.Dense(units=hidden_dim // 2, activation=tf.nn.relu)

    def call(self, input_features):
        activation = self.hidden_layer1(input_features)
        activation = self.hidden_layer2(activation)
        activation = self.hidden_layer3(activation)
        return activation
```

```python
class Decoder(K.layers.Layer):
    def __init__(self, hidden_dim, original_dim):
        super(Decoder, self).__init__()
        self.hidden_layer1 = K.layers.Dense(units=hidden_dim, activation=tf.nn.relu)
        self.hidden_layer2 = K.layers.Dense(units=hidden_dim*2, activation=tf.nn.relu)
        self.output_layer = K.layers.Dense(units=original_dim, activation=tf.nn.relu)

    def call(self, encoded):
        activation = self.hidden_layer1(encoded)
        activation = self.hidden_layer2(activation)
        activation = self.output_layer(activation)
        return activation
```

```python
class Autoencoder(K.Model):
    def __init__(self, hidden_dim, original_dim):
        super(Autoencoder, self).__init__()
        self.loss = []  # Initialize a list to store loss values during training
        self.encoder = Encoder(hidden_dim=hidden_dim)  # Initialize the Encoder
        self.decoder = Decoder(hidden_dim=hidden_dim, original_dim=original_dim)  # Initialize the Decoder

    def call(self, input_features):
        encoded = self.encoder(input_features)  # Encode input features
        reconstructed = self.decoder(encoded)  # Reconstruct encoded features
        return reconstructed
```

```python
# Instantiate the Autoencoder model
model = Autoencoder(hidden_dim=hidden_dim, original_dim=original_dim)
```

```python
 # Compile the model with mean squared error loss and optimizer Adam
model.compile(loss='mse', optimizer='adam')  #
```

```python


loss = model.fit(x_train_noisy,  # Train the model on noisy training data
                x_train,  # Target is original training data
                validation_data=(x_test_noisy, x_test),  # Validate on noisy test data
                epochs=max_epochs,  # Train for 20 epochs
                batch_size=batch_size)  # Utilize a defined batch size
plt.plot(range(max_epochs), loss.history['loss'])  # Plot training loss over epochs
plt.xlabel('Epochs')  # Label x-axis
plt.ylabel('Loss')  # Label y-axis
plt.show()  # Display the plot
```

```python
number = 10  # Number of digits to display
plt.figure(figsize=(20, 4))  # Set figure size for display
for index in range(number):
    # Display original images
    ax = plt.subplot(2, number, index + 1)
    plt.imshow(x_test_noisy[index].reshape(28, 28), cmap='gray')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)

    # Display reconstructed images
    ax = plt.subplot(2, number, index + 1 + number)
    plt.imshow(model(x_test_noisy)[index].numpy().reshape(28, 28), cmap='gray')
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
plt.show()  # Display the plot
```
